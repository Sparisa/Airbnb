```python
df = pd.read_csv("TorontoListings.csv")
```

```python
df = df[['neighbourhood',  'room_type','price']]
cols = list(df.columns.values)
print(cols)
```

```python
print(df)
```

```python
def priceRanker(val):
    if  val< 100:
        return 1
    elif 200 < val:
         return 2
    else :
         return 3
        

df["categoryPrice"] = df["price"].apply(priceRanker)
```

```python
print(df)
```

```python
import seaborn as sns
plt.figure(figsize=(15,12))
sns.scatterplot(x='room_type', y='price', data=df)

plt.xlabel("Room Type", size=13)
plt.ylabel("Price", size=13)
plt.title("Room Type vs Price",size=15, weight='bold')
```

```python
LowPrice=df[(df['categoryPrice'] == 1)]
plt.figure(figsize=(15,12))
sns.scatterplot(x='neighbourhood', y= 'price', data=LowPrice)
plt.xlabel("Neighbourhood", size=13)
plt.ylabel("Price", size=13)
plt.title("Neighbourhood vs LowPrice",size=15, weight='bold')
```

```python
MedPrice=df[(df['categoryPrice'] == 2)]
plt.figure(figsize=(15,12))
sns.scatterplot(x='neighbourhood', y= 'price', data=MedPrice)
plt.xlabel("Neighbourhood", size=13)
plt.ylabel("Price", size=13)
plt.title("Neighbourhood vs Price",size=15, weight='bold')
```

```python
HighPrice=df[df['categoryPrice'] == 3]
plt.figure(figsize=(15,12))
sns.scatterplot(x='neighbourhood', y= 'price', data=HighPrice)
plt.xlabel("Neighbourhood", size=13)
plt.ylabel("Price", size=13)
plt.title("Neighbourhood vs Price",size=10, weight='bold')
```

```python
plt.figure(figsize=(20,15))
sns.scatterplot(x="room_type", y="price",
            hue="neighbourhood", size="neighbourhood",
            sizes=(50, 200), palette="Dark2", data=df)
plt.legend(bbox_to_anchor=(1.01, 1),borderaxespad=0)
plt.xlabel("Room Type", size=13)
plt.ylabel("Price", size=13)
plt.title("Room Type vs Price vs Neighbourhood Group",size=15, weight='bold')
```
